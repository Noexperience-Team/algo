## Tri fusion

```python
def tri_fusion(w):
    def triFussion(v):

        if len(v) > 1:
            mid = len(v)//2
            left = v[:mid]
            right = v[mid:]

            triFussion(left)
            triFussion(right)

            i = 0
            j = 0
            k = 0
            while i < len(left) and j < len(right):
                if left[i] <= right[j]:
                    v[k] = left[i]
                    i = i+1
                else:
                    v[k] = right[j]
                    j = j+1
                k = k+1

            while i < len(left):
                v[k] = left[i]
                i = i+1
                k = k+1

            while j < len(right):
                v[k] = right[j]
                j = j+1
                k = k+1

    triFussion(w)
    return w
```

## Tri rapide

```python
def trirapide(L):
    """trirapide(L): tri rapide (quicksort) de la liste L"""
    def trirap(L, g, d):
        pivot = L[(g+d)//2]
        i = g
        j = d
        while True:
            while L[i] < pivot:
                i += 1
            while L[j] > pivot:
                j -= 1
            if i > j:
                break
            if i < j:
                L[i], L[j] = L[j], L[i]
            i += 1
            j -= 1
        if g < j:
            trirap(L, g, j)
        if i < d:
            trirap(L, i, d)

    g = 0
    d = len(L)-1
    trirap(L, g, d)
```

#### Creation de fonction qui permer de remplire un tableau de taille n par des entier aleatoirement

```python
def remplir(n):
    V = []
    for i in range(0, n):
        V.append(random.randint(0, 100000))

    return V
```

```python
for i in N:
    V = remplir(i)
    t = int(round(time.time() * 1000))
    v = trirapide(V)
    t1 = int(round(time.time()*1000))

    x.append(i)
    y.append(t1-t)
for i in range(len(x)):
    print('pour trie un tableau de taille {} on a besoin de {} millisecondes'.format(x[i],y[i]))
plt.plot(x, y)
plt.title('Tri Rapide')
plt.xlabel('N:taille de tableau')
plt.ylabel("T:temps d'execution")
plt.show()
```
